In this portfolio project I'll use data.csv file which contains the breast cancer dataset data. My aim is to classify the cancer type as malignant(M) or benign(B). We have to classes so it's a binary classification. Also I'll use logistic regression algorithm in my model.

```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# get the data
cancer_df = pd.read_csv('data.csv')
cancer_df.head()
```

In our dataset there is id and Unnamed: 32 column which I find unnecessary so let's drop them then start  to analyze the dataset.


```python
cancer_df.drop(['Unnamed: 32', 'id'], axis=1, inplace=True)
cancer_df.info()
```

```python
# is there any null value? if there is/are how many? (there is no null value )
cancer_df.isna().sum()
```

```python
# split the data

from sklearn.model_selection import train_test_split

X = cancer_df.drop('diagnosis', axis=1)
y = cancer_df['diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# apply standard scaling to our features to aviod any kind of warning 

from sklearn.preprocessing import StandardScaler

scl = StandardScaler()

X_train = scl.fit_transform(X_train)
X_test = scl.fit_transform(X_test)
```

```python
# generate the logistic regression model

from sklearn.linear_model import LogisticRegression

lorm = LogisticRegression()
lorm.fit(X_train, y_train)
```

```python
# make predictions
preds = lorm.predict(X_test)
```

```python
# and check the results

from sklearn import metrics

print("Accuracy Score:\n", metrics.accuracy_score(y_test, preds))
print("\nThe Confusion Matrix:\n", metrics.confusion_matrix(y_test, preds))
```

Result:  We got .99 accuracy score. And the confusion matrix certifies the accuracy score becasue there is only one misestimation.
